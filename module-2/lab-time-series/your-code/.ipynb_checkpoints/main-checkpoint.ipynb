# Challenge 2 - Modelling the Data

To model our data, we should look at a few assumptions. First, let's plot the `lag_plot` to detect any autocorrelation. Do this for `userAcceleration.x`

```python
# Your code here:
pd.plotting.lag_plot(sensor['userAcceleration.x'],lag=1)
```

Create a lag plot for `rotationRate.x`

```python
# Your code here:
pd.plotting.lag_plot(sensor['rotationRate.x'],lag=1)
```

What are your conclusions from both visualizations?

- userAcceleration.x and rotationRate.x are higly autocorrelated (they a high correlation to their lag) as the shape of the scatter to their lag approaches a 45 degree linear graph.

- rotationRate seams more autocorrelated than userAcceleration.x as the shape of the scatter seams more linear.

The next step will be to test both variables for stationarity. Perform the Augmented Dickey Fuller test on both variables below.

```python
# Your code here:
from statsmodels.tsa.stattools import adfuller as adolf

# Acceleration H0: Acceleration is not stationary
# Acceleration H1: Acceleration is stationary

# Rotation H0:Rotation is not stationary
# Rotation H0:Rotation is stationary

alpha = 0.05

adolf_acceleration = adolf(sensor['userAcceleration.x'],regression = 'ctt')
adolf_rotation = adolf(sensor['rotationRate.x'],regression = 'ctt')
print(adolf_acceleration,"\n",adolf_rotation)
```

What are your conclusions from this test?

```python
# Your conclusions here:

if adolf_acceleration[1] <alpha:
    print('H0 is rejected, Acceleration is stationary')
else:
    print('H0 is not rejected, Acceleration could be not stationary')

if adolf_rotation[1] <alpha:
    print('H0 is rejected, Rotation is stationary')
else:
    print('H0 is not rejected,  Rotation could be not stationary')
```

Finally, we'll create an __ARIMA model for `userAcceleration.x`. Load the `ARMA` function from `statsmodels`. The order of the model is (2, 1). Split the data to train and test__. Use the last 10 observations as the test set and all other observations as the training set. 

```python
# Your code here: ARIMA
# 3er valor differencing poner 2,0,1
from statsmodels.tsa.arima.model import ARIMA

test_set = sensor['userAcceleration.x'][-10:]
training_set = sensor['userAcceleration.x']

history=[ x for x in training_set]


sensor.train = ARIMA(training_set,order= (2,0,1))
sensor.test = ARIMA(test_set,order= (2,0,1))

model_fit = sensor.train.fit()
print(model_fit.summary())
```

To compare our predictions with the observed data, we can compute the RMSE (Root Mean Squared Error) from the submodule `statsmodels.tools.eval_measures`. You can read more about this function [here](https://www.statsmodels.org/dev/generated/statsmodels.tools.eval_measures.rmse.html). Compute the RMSE for the last 10 rows of the data by comparing the observed and predicted data for the `userAcceleration.x` column.

```python
# Your code here:
from statsmodels.tools import eval_measures
predict_test = model_fit.predict(test_set.index[0], test_set.index[-1], dynamic = True)
print('RMSE:',eval_measures.rmse(predict_test,test_set))
```
